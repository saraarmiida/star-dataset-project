```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./data.csv')
```

```python
data_dc = {
    'temperature': data['Temperature (K)'],
    'luminosity': data['Luminosity(L/Lo)'],
    'radius': data['Radius(R/Ro)'],
    'magnitude': data['Absolute magnitude(Mv)'],
    'star_type': data['Star type'],
    'star_color': data['Star color'],
    'spectral_class': data['Spectral Class'],
}
```

```python
R = data[data["Star type"] == 0]
B = data[data["Star type"] == 1]
W = data[data["Star type"] == 2]
M = data[data["Star type"] == 3]
S = data[data["Star type"] == 4]
H = data[data["Star type"] == 5]
```

```python
sO = data[data["Spectral Class"] == 'O']
sB = data[data["Spectral Class"] == 'B']
sA = data[data["Spectral Class"] == 'A']
sF = data[data["Spectral Class"] == 'F']
sG = data[data["Spectral Class"] == 'G']
sK = data[data["Spectral Class"] == 'K']
sM = data[data["Spectral Class"] == 'M']
```

```python
fig , ax = plt.subplots(figsize = (13,10))

ax.scatter(np.log(R["Temperature (K)"]), np.log(R["Luminosity(L/Lo)"]),15, label = 'Red dwarfs')
ax.scatter(np.log(B["Temperature (K)"]), np.log(B["Luminosity(L/Lo)"]),15, label = 'Brown dwarfs')
ax.scatter(np.log(W["Temperature (K)"]), np.log(W["Luminosity(L/Lo)"]),15, label = 'White dwarfs')
ax.scatter(np.log(M["Temperature (K)"]), np.log(M["Luminosity(L/Lo)"]),15, label = 'Main sequence stars')
ax.scatter(np.log(S["Temperature (K)"]), np.log(S["Luminosity(L/Lo)"]),15, label = 'Supergiants')
ax.scatter(np.log(H["Temperature (K)"]), np.log(H["Luminosity(L/Lo)"]),15, label = 'Hypergiants')

ax.invert_xaxis()
ax.legend()
plt.xlabel("Temperature (K)")
plt.ylabel("Luminosity (L/Lo)")
ax.grid()
```

```python
fig , ax = plt.subplots(figsize = (13,10))

ax.scatter(np.log(R["Temperature (K)"]), np.log(R["Absolute magnitude(Mv)"]),15, label = 'Red dwarfs')
ax.scatter(np.log(B["Temperature (K)"]), np.log(B["Absolute magnitude(Mv)"]),15, label = 'Brown dwarfs')
ax.scatter(np.log(W["Temperature (K)"]), np.log(W["Absolute magnitude(Mv)"]),15, label = 'White dwarfs')
ax.scatter(np.log(M["Temperature (K)"]), np.log(M["Absolute magnitude(Mv)"]),15, label = 'Main sequence stars')
ax.scatter(np.log(S["Temperature (K)"]), np.log(S["Absolute magnitude(Mv)"]),15, label = 'Supergiants')
ax.scatter(np.log(H["Temperature (K)"]), np.log(H["Absolute magnitude(Mv)"]),15, label = 'Hypergiants')

ax.invert_xaxis()
ax.legend()
plt.xlabel("Temperature (K)")
plt.ylabel("Absolute magnitude(Mv)")
ax.grid()
```
